```python
# Calculate the count of each unique value in 'Industry' and 'Work_Location'
industry_count = data['Industry'].value_counts().reset_index()
work_location_count = data['Work_Location'].value_counts().reset_index()

# Rename the columns to make them clear
industry_count.columns = ['Industry', 'Industry_Count']
work_location_count.columns = ['Work_Location', 'Work_Location_Count']

# Merge the two DataFrames based on a common index or keep them separate as needed
jobtype = pd.merge(industry_count, work_location_count, left_index=True, right_index=True, how='outer')

# Display the final DataFrame
jobtype
```

# **Diffrence between the two types of Remote Work And Industries**

# **Plot-1 bar graph**

```python

# Create a figure with two subplots (one for Industry and one for Work Location)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot bar graph for 'Industry' counts
axes[0].bar(industry_count['Industry'], industry_count['Industry_Count'], color='skyblue')
axes[0].set_title('Industry Count')
axes[0].set_xlabel('Industry')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=90)  # Rotate x-axis labels for better readability

# Plot bar graph for 'Work_Location' counts
axes[1].bar(work_location_count['Work_Location'], work_location_count['Work_Location_Count'], color='lightgreen')
axes[1].set_title('Work Location Count')
axes[1].set_xlabel('Work Location')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=90)  # Rotate x-axis labels for better readability

# Display the plots
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

**Plot-2 pie chart -Industry and 'Work_Location**

#**Inference:**
1. This pie chart provides a detailed breakdown of the distribution of industries and their respective work locations.
2. Remote work locations are emphasized using an explode effect, highlighting their proportion compared to other work setups.
3. The chart can help identify which industries have a higher tendency for remote work or specific work setups.


```python
# Merge 'Industry' and 'Work_Location' counts into a single DataFrame
industry_work_location = data.groupby(['Industry', 'Work_Location']).size().reset_index(name='Count')

# Combine the labels for pie chart (Industry and Work_Location)
industry_work_location['Label'] = industry_work_location['Industry'] + " - " + industry_work_location['Work_Location']

# Prepare the counts and labels for the pie chart
counts = industry_work_location['Count']
labels = industry_work_location['Label']

# Create an explode effect for each Work_Location to highlight the slices
explode = [0.1 if 'Remote' in label else 0 for label in labels]  # Explode remote work locations

# Create a pie chart
plt.figure(figsize=(12, 8))
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode, colors=plt.cm.Paired.colors)
plt.title('Industry and Work Location Distribution')
plt.show()
```

# **Plot 3: Bar Graph - Stress Level vs Industry**

**Inference of Stress Levels:**

1. The stacked bar chart effectively shows the proportion of different stress levels within each industry.
2. Industries with higher proportions of severe stress levels (e.g., "High Stress") may indicate more challenging or demanding work environments.

```python

# Preprocessing data for visualizations
# Aggregating Stress Level by Industry
stress_by_industry = data.groupby('Industry')['Stress_Level'].value_counts(normalize=True).unstack(fill_value=0)

# Plot 3: Bar Graph - Stress Level by Industry
plt.figure(figsize=(12, 6))
stress_by_industry.plot(kind='bar', stacked=True, colormap='viridis', ax=plt.gca())
plt.title('Stress Levels by Industry', fontsize=16)
plt.xlabel('Industry', fontsize=14)
plt.ylabel('Proportion of Stress Level', fontsize=14)
plt.legend(title='Stress Level')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

# **Plot 4: Pie Chart-Mental Health Conditions for a Specific Industry**

**Inference:**
1. The pie chart highlights the proportion of various mental health conditions reported within the IT industry.
2. The most prominent mental health condition can be identified by the largest slice of the pie.
3. Industries with high proportions of severe conditions might require targeted interventions

```python
# Aggregating Mental Health Conditions by Industry
mental_health_by_industry = data.groupby('Industry')['Mental_Health_Condition'].value_counts().unstack(fill_value=0)

# Plot 4: Pie Chart - Mental Health Conditions for a Specific Industry (e.g., IT)
industry = 'IT'  # Example industry
mental_health_it = mental_health_by_industry.loc[industry]
plt.figure(figsize=(8, 8))
mental_health_it.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title(f'Mental Health Conditions in {industry}', fontsize=16)
plt.ylabel('')  # Remove y-axis label for better aesthetics
plt.tight_layout()
plt.show()

```

# **Plot 5: Stacked Bar Chart- Job Roles OR Region**

**Inference:**
1. The stacked bar chart visually represents the distribution of mental health conditions across job roles or regions.
2. Certain roles or regions may show higher concentrations of specific conditions, indicating areas of concern.
3. Variations in conditions suggest differences in work environments, stress levels, or access to resources.

```python
# Aggregating Job Roles by Region
jobs_by_region = data.groupby('Region')['Job_Role'].value_counts().unstack(fill_value=0)

# Plot 5: Stacked Bar Chart - Job Roles by Region
plt.figure(figsize=(14, 8))
jobs_by_region.plot(kind='bar', stacked=True, colormap='tab20', ax=plt.gca())
plt.title('Job Roles Distribution by Region', fontsize=16)
plt.xlabel('Region', fontsize=14)
plt.ylabel('Count of Job Roles', fontsize=14)
plt.legend(title='Job Role', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# **Plot 6: Bar Chart - Sleep Quality for each Work Location**

**Inference:**
1. The bar chart highlights the distribution of different sleep quality levels across work locations.
2. Work-from-home setups may show higher instances of poor or average sleep quality due to blurred work-life boundaries.
3. In-office or hybrid models might exhibit better sleep quality, indicating structured routines help improve rest.

```python
# Calculate the count of each unique value in 'Industry' and 'Work_Location'
Sleep_Quality = data['Sleep_Quality']
Work_Location = data['Work_Location']

# Merge the two DataFrames based on a common index or keep them separate as needed
jobtype = pd.merge(Sleep_Quality, Work_Location, left_index=True, right_index=True, how='outer')

# Display the final DataFrame
jobtype.head(15)

```

```python
# Count occurrences of Sleep Quality for each Work Location
sleep_quality_location_counts = jobtype.groupby('Work_Location')['Sleep_Quality'].value_counts().unstack(fill_value=0)

# Plot a grouped bar chart
plt.figure(figsize=(12, 8))
sleep_quality_location_counts.plot(kind='bar', colormap='plasma', ax=plt.gca())

# Add titles and labels
plt.title('Counts of Sleep Quality by Work Location', fontsize=16)
plt.xlabel('Work Location', fontsize=14)
plt.ylabel('Count of Sleep Quality', fontsize=14)
plt.legend(title='Sleep Quality', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Display the bar graph
plt.show()
```
